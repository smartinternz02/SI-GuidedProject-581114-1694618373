### DATA VISUALISATION

```python
import seaborn as sns
# print(sns.get_dataset_names())
dataset=sns.load_dataset('car_crashes')
dataset
```

##### 1.Barplot

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(50,15))
sns.barplot(x=dataset['abbrev'],y=dataset['total'],data=dataset)
```

##### Inference: From the above graph we can see that maximum carcrashes are from city with abbrevation ND

##### 2.LinePlot

```python
sns.lineplot(x='alcohol',y='total',data=dataset)
```

##### Inference: From the above graph we can see that in the cities with high total crashes, alcohol is also high 

##### 3. BoxPlot

```python
sns.boxplot(dataset['total'])
```

##### Inference: In the above graph there are no outliers, From this we can say that there are no cities with extremely less car crashes as wells as extremely high car crashes

##### 4. Scatter Plot

```python
sns.scatterplot(x='no_previous',y='total',data=dataset)
```

##### Inference: From the above graph we can say that in every city , car crashes are higher for the first time i.e, in cities with higher car crashes, the crashes with no previuos experience or first  time crashes are also high

##### 5.Histogram

```python
sns.histplot(data=dataset,x='total')
```

##### Inference: From the above graph we can say that in most of the cities the total car crashes are in range of (10-12) 

##### 6. HeatMap

```python
correlation_matrix = dataset.corr(numeric_only=True)
sns.heatmap(correlation_matrix,annot=True)
```

##### Inference: From the above graph, we can see the positive co-relation between total crashes and speeding,alcohol and not_distracted. so total crashes depends on speeding,alcohol and no_distraction

##### 7. Pair Plot

```python
sns.pairplot(dataset, vars=['speeding', 'alcohol', 'not_distracted'])
```

##### Inference: From the above graphs we can say that all three features alcohol,speeding,not_distracted are proportional to each other

##### 8.Kernel Density Estimation Plot

```python
sns.kdeplot(data=dataset['ins_premium'], shade=True)
plt.show()
```

##### Inference: In most of the cities the insurance premium is around 800 and 1000

##### 9.Regression Plot

```python
sns.regplot(data=dataset, x='alcohol', y='total')
```

##### Inference: From the above graph we can say that there is a linear relationship between alcohol and the total carcrashes

##### 10. Violin Plot

```python
plt.figure(figsize=(40,15))
sns.violinplot (x=dataset['ins_premium'], y=dataset['total'],data=dataset)
```

##### Inference: From the above graph we can say that there is no relation between insurance premium and total car_crashes
